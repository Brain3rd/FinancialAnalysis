```python
de = pd.read_csv("results/result_de.csv", index_col=0)
de.columns = ['DE Ratio']
de.shape
```

```python
current = pd.read_csv("results/result_cr.csv", index_col=0)
current.columns = ['Current Ratio']
current.shape
```

```python
roi = pd.read_csv("results/result_roi.csv", index_col=0)
roi.columns = ['ROI']
roi.shape
```

```python
rev = pd.read_csv("results/result_rev.csv", index_col=0)
rev.set_index('company', inplace=True)
rev.columns = ['Revenue CAGR']
rev.shape
```

```python
eps = pd.read_csv("results/result_eps.csv", index_col=0)
eps.set_index('company', inplace=True)
eps.columns = ['EPS CAGR']
eps.shape
```

```python
bvps = pd.read_csv("results/result_bvps.csv", index_col=0)
# bvps.set_index('company', inplace=True)
# bvps.columns = ['BVPS']
bvps.shape
```

```python
fcf = pd.read_csv("results/result_fcf.csv", index_col=0)
fcf.set_index('company', inplace=True)
fcf.columns = ['FCF CAGR']
fcf.shape
```

```python
pe_current = pd.read_csv("results/current_pe_ratio.csv", index_col=0)
# pe_current.columns = ['Current PE']
pe_current.shape
```

```python
eps_current = pd.read_csv("results/current_eps.csv", index_col=0)
eps_current.shape
```

```python
data = pd.concat([de, current, roi, rev, eps, bvps, fcf, eps_current, pe_current], axis=1)
```

```python
points = data.sort_values('DE Ratio', ascending=False)
points = points.dropna(subset=['DE Ratio'])
# points

point = 1
for i, row in points.iterrows():
    data.at[i, 'Points'] = point
    point += 1
```

```python
points = data.sort_values('Current Ratio')
points = points.dropna(subset=['Current Ratio'])
points

point = 1
for i, row in points.iterrows():
    points = row['Points']
#     print(points)
    data.at[i, 'Points'] = point + points
    point += 1
```
