## Update current data

```python
from yahoofinancials import YahooFinancials
from http.client import IncompleteRead
import pandas as pd
import csv
from time import sleep
```

```python
with open('companies/FIN.txt') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    FIN=[i for row in data for i in row]
print(len(FIN))
```

### Earnings/Share (EPS)

```python
def get_current_eps():
    current_eps = {}
    max_try = 5
    for n in range(len(FIN)):
        attemp = 0
        while attemp < max_try:
            try:
                yf = YahooFinancials(FIN[n])
                eps_data = yf.get_earnings_per_share()
                current_eps[FIN[n]] = eps_data
                print(f"FIN_{n}")
            except IncompleteRead:
                attemp += 1
#                 sleep(10)
                if attemp == max_try:
                    print(f'Failed {N}')
            except ValueError as e:
                print(f'{e}: {N}')
                attemp = max_try
            else:
                break
                                
    df_eps = pd.DataFrame.from_dict(current_eps, orient='index')
    df_eps.columns = ['Current EPS']
    df_eps.to_csv("results/current_eps.csv")   
```

```python
get_current_eps()
```
